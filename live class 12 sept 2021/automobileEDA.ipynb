Data Exploration

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.size'] = 13.0
```

```python
#Plotting a Histogram for the symboling column
automobile['symboling'].plot(kind='hist',bins=20)
#automobile['symboling'].value_counts().plot(kind='bar',align='center',rot=1)
plt.title('Symboling - Distribution')
plt.ylabel('Count')
plt.xlabel('Symboling - Insurance Risk Factor')

plt.show()
```

```python
#Plotting a pie chart for the make attribute
automobile['make'].value_counts().plot(kind='pie',autopct='%2.0f%%',figsize=(18,18))
plt.title('Distribution Of Vehicle By Makes')
plt.show()
```

```python
#plotting a histogram for the price of vehicles from the dataset
automobile['price'].plot(kind='hist',bins=8)
plt.title('Vehicle Price- Distribution') 
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

```python
#plotting a stacked bar chart for the body-style attribute and symboling attribute
automobile.groupby(['body-style','symboling']).size().unstack().plot(kind='bar',stacked=True)
plt.ylabel('Number of Vehicles')
plt.title("Risk with body style")
plt.show()
```

```python
automobile[automobile['body-style']=='sedan']
```

```python
#plotting regplot against horsepower and price , to find the influence of hoursepower in price
import seaborn as sns
sns.regplot(x="horsepower", y="price", data=automobile);
plt.ylim(0,)
plt.title('Variation of Price with Horsepower')
plt.show()
```

```python
#plotting boxplot for drive-wheel and price variables
sns.boxplot(x="drive-wheels", y="price", data=automobile)
plt.title('Price variation on drive-wheels')
#automobile.boxplot(column='drive-wheels',by='price') 
plt.show()
```

```python
#scatter plot for all the numerical variables in the dataset
from pandas.tools.plotting import scatter_matrix 
scatter_matrix(automobile[['normalized-losses','wheel-base','length',
                          'width','height','curb-weight','engine-size',
                          'bore','stroke','compression-ratio','horsepower',
                          'peak-rpm','city-mpg','highway-mpg','price']],alpha=0.2,figsize=(19,20),diagonal='hist') 
plt.show()
```
